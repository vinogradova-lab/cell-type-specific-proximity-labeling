```python
import pandas as pd 
```

```python
mouse_tfs_df = pd.read_csv("./mouse/01_cyt_nuc_TFs/mouse_proteome_TFs.csv", index_col=0)
mouse_cytoskeletal_df = pd.read_csv("./mouse/01_cyt_nuc_TFs/uniprot-compressed_cytoskeletal_mouse.csv")
mouse_nucleus_df = pd.read_csv("./mouse/01_cyt_nuc_TFs/uniprot-compressed_nucleus_MS.csv")
```

**Combine Cyt, Nuc and TF files**

```python
df_to_combine_list = [mouse_tfs_df, mouse_cytoskeletal_df, mouse_nucleus_df]
```

```python
for df in df_to_combine_list:
    print(df.shape)
    
for df in df_to_combine_list:
    print(df.columns)
```

```python
col_list = ['Entry', 'Reviewed', 'Entry Name', 'Protein names', 'Gene Names',
       'Organism', 'Length', 'Gene Ontology (cellular component)',
       'Subcellular location [CC]', 'Mass', 'Function [CC]',
       'Tissue specificity', 'Gene Ontology (biological process)',
       'Gene Ontology (molecular function)', 'Signal peptide',
       'Transit peptide']

cyt_nuc_df = mouse_cytoskeletal_df.merge(mouse_nucleus_df, on=col_list, how='outer')
cyt_nuc_df.shape
```

```python
col_list = ['Entry', 'Reviewed', 'Entry Name', 'Protein names', 'Gene Names',
       'Organism', 'Length', 'Gene Ontology (cellular component)',
       'Gene Ontology (biological process)',
       'Gene Ontology (molecular function)']

cyt_nuc_tfs_df = cyt_nuc_df.merge(mouse_tfs_df, on=col_list, how='outer')
cyt_nuc_tfs_df.shape
```

**Combine secreted lists**

```python
mouse_secreted_df = pd.read_csv("./mouse/02_secreted/uniprot-compressed_secreted_mouse_08032022.csv")
mouse_secreted_es_df = pd.read_csv("./mouse/02_secreted/uniprot-compressed_Mouse_Extracellular space.csv")
mouse_secreted_ee_df = pd.read_csv("./mouse/02_secreted/uniprot-compressed_Mouse_Extracellular exosome.csv")
mouse_secreted_human_atlas_df = pd.read_csv("./mouse/02_secreted/uniprot-download_true_fields_accession_2Creviewed_2Cid_2Cprotein_nam-2022.08.11-22.05.45.96_mouse_human_atlas.csv")
```

```python
mouse_secreted_to_combine_list = [mouse_secreted_df, mouse_secreted_es_df, mouse_secreted_ee_df, mouse_secreted_human_atlas_df]
```

```python
mouse_secreted_ee_df["DNA binding"] = mouse_secreted_ee_df["DNA binding"].astype(str)
mouse_secreted_es_df["Transit peptide"] = mouse_secreted_es_df["Transit peptide"].astype(str)
```

```python
for df in mouse_secreted_to_combine_list:
    print(df.shape)
    
for df in mouse_secreted_to_combine_list:
    print(df.columns)
```

```python
col_list = ['Entry', 'Reviewed', 'Entry Name', 'Protein names', 'Gene Names',
       'Organism', 'Length', 'Gene Ontology (cellular component)',
       'Subcellular location [CC]', 'Mass', 'Function [CC]',
       'Tissue specificity', 'Gene Ontology (biological process)',
       'Gene Ontology (molecular function)', 'Signal peptide',
       'Transit peptide', 'DNA binding', 'Gene Ontology (GO)',
       'Protein families', 'Zinc finger']

es_ee_df = mouse_secreted_es_df.merge(mouse_secreted_ee_df, on=col_list, how='outer')
es_ee_df.shape
```

```python
col_list = ['Entry', 'Reviewed', 'Entry Name', 'Protein names', 'Gene Names',
       'Organism', 'Length', 'Gene Ontology (cellular component)',
       'Subcellular location [CC]', 'Function [CC]', 'Mass',
       'Tissue specificity', 'Gene Ontology (biological process)',
       'Signal peptide', 'Transit peptide']

secreted_es_ee_df = es_ee_df.merge(mouse_secreted_df, on=col_list, how='outer')
secreted_es_ee_df.shape
```

```python
col_list = ['Entry', 'Reviewed', 'Entry Name', 'Protein names', 'Gene Names',
       'Organism', 'Length', 'Gene Ontology (biological process)',
       'Gene Ontology (GO)', 
       'Gene Ontology (cellular component)',
       'Gene Ontology (molecular function)']

secreted_es_ee_atlas_df = secreted_es_ee_df.merge(mouse_secreted_human_atlas_df, on=col_list, how='outer')
secreted_es_ee_atlas_df.shape
```

```python
secreted_es_ee_atlas_df.to_csv("./mouse/00_combined/mouse_TP_list_secreted.csv", index=False)
```

**Remove overlapping uniprot IDs with secreted df**

```python
secreted_uniprot_list = secreted_es_ee_atlas_df["Entry"].tolist()
len(secreted_uniprot_list)
```

```python
unique_cyt_nuc_tfs_df = cyt_nuc_tfs_df[~cyt_nuc_tfs_df["Entry"].isin(secreted_uniprot_list)]
```

```python
unique_cyt_nuc_tfs_df.shape
```
