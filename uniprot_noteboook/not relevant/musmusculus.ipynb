```python
import pandas as pd
```

```python
mus_df = pd.read_csv("musmus_uniprot_main.csv")
gene_list_df = pd.read_csv("TFs.csv")
#gene_list = gene_list_df["TFs (GSEA, 2022)"].tolist()
#len(gene_list)

#for item in gene_list:
#    print(item+", ")

mus_df.shape
#gene_list_df.shape
```

```python
gene_list = gene_list_df["TFs (GSEA, 2022)"].tolist()
```

```python
def check_if_contains(gene_names):
    list_gene_names = gene_names.split()
    #list_gene_names = list_gene_names
    result_list = []
    for name in list_gene_names:
        if name.upper() in gene_list:
            result_list.append(name.upper())
    return("".join(result_list))
```

```python
mus_df["TFs (GSEA, 2022)"] = mus_df["Gene Names"].apply(check_if_contains)
mus_df
```

```python
merged_df = pd.merge(mus_df, gene_list_df, on="TFs (GSEA, 2022)", how="outer")
```

```python
merged_df.to_csv("mus_musc_merged_df.csv")
```
