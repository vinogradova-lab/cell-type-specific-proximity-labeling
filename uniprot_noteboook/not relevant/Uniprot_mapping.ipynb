```python
import pandas as pd
import urllib.parse
import urllib.request
import io
```

```python
gene_list_df = pd.read_csv("TFs.csv")
gene_list = gene_list_df["TFs (GSEA, 2022)"].tolist()
len(gene_list)
```

```python
url = 'https://legacy.uniprot.org/uploadlists/'

params = {
    'from': 'GENENAME',
    'to': 'ACC',
    'format': 'tab',
    'query': " ".join(gene_list),
    }
```

```python
data = urllib.parse.urlencode(params)
data = data.encode('utf-8')
req = urllib.request.Request(url, data)

with urllib.request.urlopen(req) as f:
    response = f.read()
    # result = response.decode('utf-8')
    # print(result)
df_fasta = pd.read_csv(io.StringIO(response.decode("utf-8")), sep="\t")
    #print(df_fasta)
#df_fasta.columns = ["GO (cellular component)", "Subcellular location", "uniprot"]
#df_fasta["Subcellular location"] = df_fasta["Subcellular location"].str.replace("SUBCELLULAR LOCATION:", "")  

df_fasta
```
