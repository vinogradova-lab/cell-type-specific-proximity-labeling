```python
manual_results_df = pd.read_csv("biomart_uniprot_manual.csv")
```

```python
manual_results_df
list_manual_results = manual_results_df["gene_name"].tolist()
```

```python
uniprot_table = pd.read_csv("uniprot_table.csv")
uniprot_table.columns
uniprot_table.shape
```

```python
def check_if_contains(gene_names):
    list_gene_names = gene_names.split()
    #list_gene_names = list_gene_names
    result_list = []
    for name in list_gene_names:
        if name in list_manual_results:
            result_list.append(name)
    return("".join(result_list))
```

```python
uniprot_table["TFs (GSEA, 2022)"] = uniprot_table["Gene Names"].apply(check_if_contains)
#uniprot_table[["Gene Names", "TFs (GSEA, 2022)"]]

```

```python
uniprot_table.shape
```

```python
manual_results_df.columns = ['TFs (GSEA, 2022)', 'uniprot_id', 'note']
```

```python
uniprot_table = pd.merge(uniprot_table, manual_results_df, on="TFs (GSEA, 2022)", how="left")
uniprot_table = uniprot_table.drop_duplicates()
uniprot_table
```
