```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#import TF names as list
gene_list_df = pd.read_csv("TFs.csv")
gene_list = gene_list_df["TFs (GSEA, 2022)"].tolist()
```

```python
#import mouse proteome table
mouse_proteome_df = pd.read_csv("./mouse_uniprot_protein_table/uniprot-download_true_fields_accession_2Creviewed_2Cid_2Cprotein_nam-2022.08.08-13.40.09.29.csv")
```

```python
def check_if_contains(gene_names):
    gene_names = str(gene_names) #.split(" ")
    gene_names_row_list = gene_names.split()
    result_list = []
    for name in gene_names_row_list:
        if name.upper() in gene_list:
            result_list.append(name.upper())
    
    if result_list == []:
        result = np.nan
    else:
        result = ", ".join(result_list)
    return(result)
    #return gene_names_row_list
```

```python
mouse_proteome_df["label"] = mouse_proteome_df["Gene Names"].apply(check_if_contains)
mouse_proteome_df.shape
```

```python
mouse_proteome_df = mouse_proteome_df[mouse_proteome_df['label'].notna()]
mouse_proteome_df.shape
```

```python
mouse_proteome_df[['label', 'second_label']] = mouse_proteome_df["label"].str.split(', ', expand=True)
```

```python
list_of_first_labels = mouse_proteome_df['label'].dropna().drop_duplicates().tolist()
len(list_of_first_labels)
```

```python
list_of_second_labels = mouse_proteome_df['second_label'].dropna().drop_duplicates().tolist()
len(list_of_second_labels)
```

```python
not_found_in_first_list = []
for name in list_of_second_labels:
    if name not in list_of_first_labels:
        not_found_in_first_list.append(name)
```

```python
not_found_in_first_list
```

```python
for item in not_found_in_first_list:
    list_of_first_labels.append(item)
```

```python
len(list_of_first_labels)
```

```python
#Check which ones were not found
```

```python
not_found_list = []
for name in gene_list:
    if name not in list_of_first_labels:
        not_found_list.append(name)
```

```python
len(not_found_list)
```

```python
not_found_df = pd.DataFrame(not_found_list)
```

```python
not_found_df.columns = ["not_found"]
```

```python
mouse_proteome_df = mouse_proteome_df.append(not_found_df)
mouse_proteome_df.shape
```

```python
mouse_proteome_df.to_csv("./mouse_uniprot_protein_table/mouse_proteome_TFs.csv")
```
