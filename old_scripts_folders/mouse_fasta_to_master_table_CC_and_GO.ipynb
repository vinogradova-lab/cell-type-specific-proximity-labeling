```python
from Bio import SeqIO
import pandas as pd
```

```python
mouse_fasta_path = "./fasta_files/radus_database__cravattlab_mouse_nonredundant_07-11-2017_reversed.fasta"
```

```python
fasta_dict = {}
with open(mouse_fasta_path) as handle:
    for record in SeqIO.parse(handle, "fasta"):
        fasta_dict[record.id] = record.description
```

```python
df = pd.DataFrame.from_dict(fasta_dict, orient='index')
```

```python
df = df.reset_index()
```

```python
df.columns = ["uniprot_id", "description"]
```

```python
df["id"] = df["uniprot_id"]
```

```python
df.head()
```

```python
df[["db","uniprot_id","entry_name"]] = df["uniprot_id"].str.split("|", expand=True)
```

```python
df.head()
```

```python
df["reverse"] = df["id"].str.contains("Reverse_")
```

```python
df.shape
```

```python
df = df[df["reverse"] == False]
```

```python
master_table = df.drop("reverse", axis=1)
```

```python
master_table.shape
```

```python
master_table.set_index("id", inplace=True)
```

```python
master_table.to_csv("mouse_master_table.csv") #run uniprot ID mapping 
```

```python
#import uniprot id mapping results 
uniprot_id_mapping_path = "./fasta_files/uniprot_id_mapping/mouse_mapping_uniprot-download_true_fields_accession_2Creviewed_2Cid_2Cprotein_nam-2022.08.31-15.17.09.71.csv"
```

```python
def annotate_TP(subcelullar_location):
    tp_term_list = ["Endoplasmic Reticulum", "Golgi" , "golgi", "Secreted", "secreted", "Endoplasmic reticulum", "Rough endoplasmic reticulum", "Extracellular space", "extracellular space", "endoplasmic reticulum"]
    subcelullar_location = str(subcelullar_location)
    if "Note:" in subcelullar_location:
        subcelullar_location = subcelullar_location.split("Note:")[0]
    for item in tp_term_list: 
        if item in subcelullar_location:
    #if any(ext in subcelullar_location for ext in tp_term_list):
            return("TP")
    
    
def annotate_FP(subcelullar_location):
    fp_term_list = ["Cytoskeleton", "Nucleus", "nucleus", 'cytoskeleton']
    subcelullar_location = str(subcelullar_location)
    if "Note:" in subcelullar_location:
        subcelullar_location = subcelullar_location.split("Note:")[0]
    for item in fp_term_list: 
        if item in subcelullar_location:
    #if any(ext in subcelullar_location for ext in fp_term_list):
            return("FP")
```

```python
#TP
merged_master["TP"] = merged_master["Subcellular location [CC]"].apply(annotate_TP) 
merged_master["TP_GO"] = merged_master["Gene Ontology (cellular component)"].apply(annotate_TP) 
merged_master["cyt_nuc"] = merged_master["Subcellular location [CC]"].apply(annotate_FP) 
merged_master["cyt_nuc_GO"] = merged_master["Gene Ontology (cellular component)"].apply(annotate_FP) #str.contains(any(tp_term_list)) #str.contains(any(tp_term_list))
```
