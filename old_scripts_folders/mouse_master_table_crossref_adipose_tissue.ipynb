```python
import pandas as pd
from pathlib import Path
import openpyxl
```

```python
master_table_path = "/Users/nropek/Dropbox (Dropbox @RU)/TurboID manuscript/Mass-spectrometry datasets/TurboID_analysis/mouse_lists/master_list/CC_GO_09022022_mouse_uniprot_mapping_master_table.csv"
tissue_localisation_path = "/Users/nropek/Dropbox (Dropbox @RU)/TurboID manuscript/Reference datasets/Mouse/Tissue localisation"
tissue_localisation_path_obj = Path(tissue_localisation_path)
```

```python
files_in_folder = list(tissue_localisation_path_obj.iterdir())
```

```python
master_table = pd.read_csv(master_table_path)
```

```python
for file_path in files_in_folder:
    column_name = file_path.stem.split("_", 1)[0]
    df = pd.read_excel(file_path)
    entry_list = df.Entry.tolist()
    master_table[column_name] = master_table.uniprot_id.isin(entry_list)
```

```python
master_table.to_csv("annCC_GO_09022022_mouse_master_with_tissue_loc_annotation.csv")
```
