```python
from Bio import SeqIO
import pandas as pd
import numpy as np
from pathlib import Path
import openpyxl
```

```python
mouse_fasta_path = "./fasta_files/radus_database__cravattlab_mouse_nonredundant_07-11-2017_reversed.fasta"
```

```python
fasta_dict = {}
with open(mouse_fasta_path) as handle:
    for record in SeqIO.parse(handle, "fasta"):
        fasta_dict[record.id] = record.description
```

```python
df = pd.DataFrame.from_dict(fasta_dict, orient='index')
```

```python
df = df.reset_index()
```

```python
df.columns = ["uniprot_id", "description"]
```

```python
df["id"] = df["uniprot_id"]
```

```python
df.head()
```

```python
df[["db","uniprot_id","entry_name"]] = df["uniprot_id"].str.split("|", expand=True)
```

```python
df.head()
```

```python
df["reverse"] = df["id"].str.contains("Reverse_")
```

```python
df.shape
```

```python
df = df[df["reverse"] == False]
```

```python
master_table = df.drop("reverse", axis=1)
```

```python
def annotate_TP(subcelullar_location):
    tp_term_list = ["Endoplasmic Reticulum", "Secreted", "secreted", "Endoplasmic reticulum", "Rough endoplasmic reticulum", "endoplasmic reticulum"]
    #"Golgi", "Extracellular space", "extracellular space", 
    #
    subcelullar_location = str(subcelullar_location)
    if "Note:" in subcelullar_location:
        subcelullar_location = subcelullar_location.split("Note:")[0]
    for item in tp_term_list: 
        if item in subcelullar_location:
            return("TP")
    
def annotate_FP(subcelullar_location):
    fp_term_list = ["Nucleus", "nucleus", "Mitochondrion", "mitochondrion"] #'cytoskeleton' "Cytoskeleton", 
    subcelullar_location = str(subcelullar_location)
    if "Note:" in subcelullar_location:
        subcelullar_location = subcelullar_location.split("Note:")[0]
    for item in fp_term_list: 
        if item in subcelullar_location:
            return("FP")
```

```python
#TP
merged_master["TP"] = merged_master["Subcellular location [CC]"].apply(annotate_TP) 
merged_master["FP"] = merged_master["Subcellular location [CC]"].apply(annotate_FP) #chnage to cyt_nuc if you add Transcirption factors
```
